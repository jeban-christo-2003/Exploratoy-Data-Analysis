```python
import pandas as pd
import random

# Define lists of car makes, models, fuel types, and transmission types
makes = ['Toyota', 'Honda', 'Ford', 'Chevrolet', 'BMW', 'Audi', 'Mercedes', 'Nissan', 'Subaru', 'Hyundai']
models = ['Camry', 'Civic', 'Mustang', 'Tahoe', '3 Series', 'A4', 'C-Class', 'Altima', 'Outback', 'Sonata', 
          'Fusion', 'Accord', 'Corolla', '5 Series', 'Q5', 'E-Class', 'Maxima', 'Highlander', 'Forester', 'Elantra',
          'Escape', 'CR-V', 'Malibu', 'X3', 'A6', 'GLC-Class', 'Rogue', 'RAV4', 'Crosstrek', 'Tucson',
          'Explorer', 'Pilot', 'Equinox', '7 Series', 'Q7', 'GLE-Class', 'Pathfinder', 'Sienna', 'Ascent', 'Santa Fe',
          'Edge', 'Odyssey', 'Traverse', 'X5', 'Q8', 'GLS-Class', 'Armada', 'Sequoia', 'Palisade', 'Expedition', 'Passport']

fuel_types = ['Gasoline', 'Diesel', 'Hybrid']
transmissions = ['Automatic', 'Manual']

# Generate 100 records
data = []
for _ in range(100):
    make = random.choice(makes)
    model = random.choice(models)
    year = random.randint(2016, 2021)
    price = random.randint(15000, 80000)
    mileage = random.randint(10000, 40000)
    fuel_type = random.choice(fuel_types)
    transmission = random.choice(transmissions)
    data.append([make, model, year, price, mileage, fuel_type, transmission])

# Create a DataFrame from the data
df = pd.DataFrame(data, columns=['Make', 'Model', 'Year', 'Price (USD)', 'Mileage (miles)', 'Fuel Type', 'Transmission'])

# Save DataFrame to CSV file
df.to_csv('car_dataset.csv', index=False)
```
